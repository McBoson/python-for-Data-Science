# object oriented programming

```python
class vehicle:
    def __init__(self, model, year, gear):
        self.model = model
        self.year = year
        self.gear = gear
    
    def movement(self):
        print(f"{self.model} {self.year} is on motion")
        
    def color(self):
        print("my color is black")
```

```python
car1= vehicle("toyota", 2014, "automatic")
car1.movement()
```

```python
car1= vehicle("benz", 2024, "automatic")
car1.movement()
```

```python
class electriccars(vehicle):
    def __init__(self, model, year, gear, battery_capacity):
        super().__init__(model, year, gear)
        self.battery_capacity= battery_capacity
    
    def charging(self):
        print(f"{self.model} is on charging")
        
```

```python
tesla = electriccars("tesla", 2019, "automatic", 200)
tesla.charging() # calling the method from the child class
tesla.movement() # calling the method from the parent class
```

```python
class gasolinecars(vehicle):
    def __init__(self, model, year, gear,tank_size):
        super().__init__(model, year, gear)
        self.tank_size=tank_size
    
    def tank(self):
        print(f"{self.model} tank is on full")
```

```python
innosen = gasolinecars("innosen", 2012, "automatic", "100kg")
innosen.tank() # calling the method from the child class
innosen.movement() # calling the method from the parent class
```

```python
class caroperation:
    def __init__(self, maker, model, year):
        self.maker = maker
        self.model = model
        self.year = year
        self.engine_on = False
        self.speed = 0
    
    def start_engine(self):
        if self.engine_on==False:
            print(f"{self.maker} {self.model}, {self.year} started!!")
            self.engine_on=True
        else:
            print(f"{self.maker} {self.model}, {self.year} already on start mode!!")
        
    def off_engine(self):
        if self.engine_on==True:
            print(f"{self.maker} {self.model}, {self.year} turn off!!")
            self.engine_on=False
        else:
            print(f"{self.maker} {self.model}, {self.year} already on off mode!!")
            
    def accelerate(self, acc):
        if self.engine_on==True:
            self.speed+=acc
            print(f"{self.maker} {self.model}, {self.year} accelerated to {self.speed} km/hr")
            if self.speed>80:
                print("speed limit exceeded, please decelerate")
                
    def decelerate(self, dcc):
        if self.speed>0:
            self.speed-=dcc
            print(f"{self.maker} {self.model}, {self.year} decelerated to {self.speed} km/hr")
            if self.speed==0:
                print("syour deceleration can not go beyound zero")
                self.speed=0
        else:
            print("you need to start the engine before acceleration")
                  
```

```python
car1 = caroperation("toyota", "camry", 2019)
```

```python
car1.off_engine()
```

```python
car1.start_engine()
```

```python
car1.accelerate(20)
```

```python
car1.decelerate(20)
```
